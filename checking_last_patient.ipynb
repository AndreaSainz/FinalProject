```python
import os
import h5py
import torch
import matplotlib.pyplot as plt
```

```python
# read one file of test data
with h5py.File('Data/ground_truth_test/ground_truth_test_000.hdf5', 'r') as f:
            ground_truth = f['data'][:]
```

```python
# Convert to tensor
sample = torch.tensor(ground_truth, dtype=torch.float32)
```

```python
sample.shape # number of photos x dimensions
```

```python
sample_image = sample[5]  

#Visualize one image
plt.imshow(sample_image, cmap='gray')
plt.colorbar()  
plt.show()
```

```python
#Visualize more than 1 image
n_images = 21  

# Create a figure to show images
fig, axes = plt.subplots(n_images // 3, 3, figsize=(n_images // 3, 9))

# loop for images
for i in range(n_images):
    ax = axes[i // 3, i % 3]  
    ax.imshow(sample[i], cmap='gray')
    ax.axis('off')  #No axis to see them clearly

plt.show()
```

## Checking number of images in last file 

```python
# read one file of test data
with h5py.File('Data/ground_truth_test/ground_truth_test_027.hdf5', 'r') as f:
            ground_truth = f['data'][:]
sample_f = torch.tensor(ground_truth, dtype=torch.float32)
sample_f.shape # number of photos x dimensions
```

```python
number_invalid = 0
for i in range(128):
    image = sample_f[i]
    if torch.all(image == 0) or torch.all(torch.isnan(image)):
        number_invalid += 1

print(f"Number of invalid images: {number_invalid}")
```
